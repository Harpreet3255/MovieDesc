```python
#Standardize videos
import os

def standardize_video(input_path, output_path, resolution="1280x720"):
    """Convert video to MP4 format and resize it to a fixed resolution."""
    os.system(f'ffmpeg -i "{input_path}" -vf scale={resolution} -c:v libx264 -preset slow -crf 22 -c:a aac "{output_path}"')
    print(f"Standardized: {output_path}")

# Standardize all videos
input_folder = "C:/Users/jagji/MovieDesc/downloaded_videos"
output_folder = "standardized_videos"
os.makedirs(output_folder, exist_ok=True)

for video_file in os.listdir(input_folder):
    input_path = os.path.join(input_folder, video_file)
    output_path = os.path.join(output_folder, video_file)  # Save with same filename
    standardize_video(input_path, output_path)
```

```python
# extract frames
import cv2
import os

def extract_frames(video_path, output_folder, frame_rate=1):
    """Extract frames from a video at the specified frame rate."""
    os.makedirs(output_folder, exist_ok=True)  # Ensure folder exists

    cap = cv2.VideoCapture(video_path)
    fps = int(cap.get(cv2.CAP_PROP_FPS))
    
    if fps == 0:
        print(f"⚠️ Warning: Unable to read FPS for {video_path}. Skipping...")
        return
    
    frame_interval = max(1, fps // frame_rate)  # Extract frame every X seconds
    frame_count = 0
    saved_frames = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_interval == 0:
            frame_filename = os.path.join(output_folder, f"{os.path.basename(video_path)}_frame_{frame_count}.jpg")
            cv2.imwrite(frame_filename, frame)
            print(f"✅ Saved frame to: {frame_filename}")
            saved_frames += 1
        frame_count += 1
    
    cap.release()
    print(f"🎉 Extraction complete: {saved_frames} frames saved from {video_path}")

# ✅ Extract frames from all downloaded videos
video_folder = "video_downloads"  # Make sure this is correct
output_folder = "extracted_frames"

if os.path.exists(video_folder):
    for video_file in os.listdir(video_folder):
        video_path = os.path.join(video_folder, video_file)
        print(f"🔄 Processing {video_file}...")
        extract_frames(video_path, output_folder)
else:
    print("❌ Error: No videos found in the folder!")
```

```python
import os
from PIL import Image

def standardize_image(image_path, output_path, size=(224, 224)):
    """Resize image to a fixed size and save it in JPG format."""
    try:
        img = Image.open(image_path).convert("RGB")  # Ensure RGB format
        img = img.resize(size, Image.Resampling.LANCZOS)  # FIX: Use LANCZOS instead of ANTIALIAS
        img.save(output_path, "JPEG")  # Save in JPG format
        print(f" Standardized: {output_path}")
    except Exception as e:
        print(f" Error processing {image_path}: {e}")

# Define folders
input_folder = "extracted_frames"  # Ensure this folder contains the images
output_folder = "standardized_frames"
os.makedirs(output_folder, exist_ok=True)  # Create folder if not exists

# Process all images
for img_file in os.listdir(input_folder):
    if img_file.endswith((".jpg", ".png", ".jpeg")):  # Process only image files
        input_path = os.path.join(input_folder, img_file)
        output_path = os.path.join(output_folder, f"standardized_{img_file}")
        standardize_image(input_path, output_path)

print(" Image standardization complete!")
```
