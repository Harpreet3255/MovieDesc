```python
import pandas as pd
import numpy as np
data=pd.read_csv('diverse_sample.csv')
data.head(5)
```

```python
import os
import requests
output_dir = 'downloaded_videos'
if not os.path.exists(output_dir):
    os.makedirs(output_dir)
# Function to download a video from a URL
def download_video(url, video_id, cluster):
    try:
        # Send GET request to the video URL
        response = requests.get(url, stream=True) 
        # Check if the request was successful
        if response.status_code == 200:
            video_path = os.path.join(output_dir, f"cluster_{cluster}_video_{video_id}.mp4")  # Save with cluster and video_id
            # Open the file in write-binary mode
            with open(video_path, 'wb') as f:
                for chunk in response.iter_content(chunk_size=1024):
                    if chunk:
                        f.write(chunk)
            print(f"Cluster {cluster} - Video {video_id} downloaded successfully.")
        else:
            print(f"Cluster {cluster} - Failed to download video {video_id}. Status code: {response.status_code}")
    except Exception as e:
        print(f"Cluster {cluster} - Error downloading video {video_id}: {str(e)}")
# Group the dataset by 'cluster'
grouped_data = data.groupby('cluster')
# Loop through each cluster and download 100 videos (if available)
for cluster, group in grouped_data:
    # Select only the first 100 videos in the current cluster
    videos_to_download = group.head(100)  # Adjust the number if needed
    for index, row in videos_to_download.iterrows():
        video_id = row['videoid']
        video_url = row['contentUrl']
        # Download the video for the current row in the cluster
        download_video(video_url, video_id, cluster)
```
